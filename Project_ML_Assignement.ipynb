# Answer to the work

### Imports

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.model_selection import GridSearchCV
```

### Model testing and fine tuning Random Forest, Logistic Regression, Support Vector Machine

```python

# Assuming X and Y are already defined
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Scale the data (optional but often recommended for logistic regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a list of models to test
models = [
    ("Random Forest", RandomForestClassifier(), {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}),
    ("Logistic Regression", LogisticRegression(max_iter=1000), {'C': [0.001, 0.01, 0.1, 1, 10, 100]}),
    ("Support Vector Machine", SVC(), {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']})
]

evaluation_results = []

# Train each model using cross-validation, hyperparameter tuning, and evaluate
for name, model, param_grid in models:
    # Perform hyperparameter tuning using GridSearchCV
    grid_search = GridSearchCV(model, param_grid, cv=10, scoring='accuracy')
    grid_search.fit(X_train_scaled, Y_train)

    # Get the best model from the grid search
    best_model = grid_search.best_estimator_

    # Perform cross-validation on the training set
    cv_scores = cross_val_score(best_model, X_train_scaled, Y_train, cv=10)

    # Make predictions on the test set
    predictions = best_model.predict(X_test_scaled)

    # Evaluate on the test set
    accuracy = accuracy_score(Y_test, predictions)
    precision = precision_score(Y_test, predictions, average='weighted')
    recall = recall_score(Y_test, predictions, average='weighted')
    f1 = f1_score(Y_test, predictions, average='weighted')

    # Use the mean of cross-validation scores
    mean_cv_score = cv_scores.mean()

    # Append the results to the list
    evaluation_results.append({
        'Model': name,
        'Best Parameters': grid_search.best_params_,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'Mean CV Score': mean_cv_score
    })

# Convert the list of dictionaries to a DataFrame using pd.concat
evaluation_results = pd.concat([pd.DataFrame(result, index=[0]) for result in evaluation_results], ignore_index=True)
```

```python
print(evaluation_results)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Plot pour Accuracy
sns.barplot(x='Model', y='Accuracy', data=evaluation_results, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Model Accuracy')

# Plot pour Precision
sns.barplot(x='Model', y='Precision', data=evaluation_results, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Model Precision')

# Plot pour Recall
sns.barplot(x='Model', y='Recall', data=evaluation_results, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Model Recall')

# Plot pour F1 Score
sns.barplot(x='Model', y='F1 Score', data=evaluation_results, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('Model F1 Score')

# Ajustement de l'espacement entre les sous-plots
plt.tight_layout()

# Affichage du graphique
plt.show()
```

```python
# Assuming X and Y are already defined
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Scale the data (optional but often recommended for logistic regression)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a list of models to bag
models = [
    ("Random Forest", RandomForestClassifier()),
    ("Logistic Regression", LogisticRegression(max_iter=1000)),
    ("Support Vector Machine", SVC())
]

evaluation_results = []

# Bagging for each model
for name, model in models:
    # Create a BaggingClassifier for the current base model
    bagging_model = BaggingClassifier(base_estimator=model, n_estimators=10, random_state=42)

    # Perform cross-validation on the training set
    cv_scores = cross_val_score(bagging_model, X_train_scaled, Y_train, cv=10)

    # Train the bagging model on the full training set
    bagging_model.fit(X_train_scaled, Y_train)

    # Make predictions on the test set
    predictions = bagging_model.predict(X_test_scaled)

    # Evaluate on the test set
    accuracy = accuracy_score(Y_test, predictions)
    precision = precision_score(Y_test, predictions, average='weighted')
    recall = recall_score(Y_test, predictions, average='weighted')
    f1 = f1_score(Y_test, predictions, average='weighted')

    # Use the mean of cross-validation scores
    mean_cv_score = cv_scores.mean()

    # Append the results to the list
    evaluation_results.append({
        'Model': f'Bagged {name}',
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'Mean CV Score': mean_cv_score
    })

# Convert the list of dictionaries to a DataFrame using pd.concat
evaluation_results = pd.concat([pd.DataFrame(result, index=[0]) for result in evaluation_results], ignore_index=True)
print(evaluation_results)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Plot pour Accuracy
sns.barplot(x='Model', y='Accuracy', data=evaluation_results, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Model Accuracy')

# Plot pour Precision
sns.barplot(x='Model', y='Precision', data=evaluation_results, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Model Precision')

# Plot pour Recall
sns.barplot(x='Model', y='Recall', data=evaluation_results, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Model Recall')

# Plot pour F1 Score
sns.barplot(x='Model', y='F1 Score', data=evaluation_results, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('Model F1 Score')

# Ajustement de l'espacement entre les sous-plots
plt.tight_layout()

# Affichage du graphique
plt.show()
```

### X_1 Y_1

```python
# Define the number of folds for cross-validation
num_folds = 10

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_1, Y_1, test_size=0.2, random_state=42)

# Create a list of models to test
models = [
    ("Random Forest", RandomForestClassifier(n_estimators=5,max_depth=5,min_samples_split=2, min_samples_leaf=2)),
    ("Logistic Regression", LogisticRegression()),
    ("Support Vector Machine", SVC())
]

evaluation_results = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])

# Cross-validation loop
for name, model in models:
    # Use StratifiedKFold for classification tasks to ensure class balance in each fold
    kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

    # Perform cross-validation
    cv_accuracy = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
    cv_precision = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='precision_weighted')
    cv_recall = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='recall_weighted')
    cv_f1 = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='f1_weighted')

    # Store the mean values of cross-validated metrics
    accuracy = cv_accuracy.mean()
    precision = cv_precision.mean()
    recall = cv_recall.mean()
    f1 = cv_f1.mean()

    evaluation_results = evaluation_results.append({
        'Model': name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    }, ignore_index=True)

print(evaluation_results)

# Plotting the results
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Plot for Accuracy
sns.barplot(x='Model', y='Accuracy', data=evaluation_results, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Model Accuracy')

# Plot for Precision
sns.barplot(x='Model', y='Precision', data=evaluation_results, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Model Precision')

# Plot for Recall
sns.barplot(x='Model', y='Recall', data=evaluation_results, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Model Recall')

# Plot for F1 Score
sns.barplot(x='Model', y='F1 Score', data=evaluation_results, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('Model F1 Score')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
